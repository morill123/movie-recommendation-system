```
# This code uses a specific API to retrieve movie information, and the size of the movie database can be self-adjusted. 
# The program sets 4000 movies. 
# If the specified movie is not found, please expand the movie database appropriately.

import requests
import pandas as pd

api_key = '9bc490e4125fa84ed4abb11b380b6ee2'
base_url = 'https://api.themoviedb.org/3'
endpoint = '/movie/top_rated'

# Set up parameters
params = {
    'api_key': api_key,
    'language': 'en-US',
    'page': 1  }

# Initialize an empty list to store movie data
all_movies = []

# Make requests to the TMDb API to get movie data
for page in range(1, 300):  # Assuming 5000 movies with 20 movies per page
    params['page'] = page
    response = requests.get(f'{base_url}{endpoint}', params=params)

    if response.status_code == 200:
        data = response.json()
        all_movies.extend(data['results'])
    else:
        print(f"Error {response.status_code}: {response.text}")

# Extract required information for each movie
movies_data = []
for movie in all_movies:
    movie_id = movie['id']

    # Get additional details for each movie
    movie_details_endpoint = f'/movie/{movie_id}'
    movie_details_response = requests.get(f'{base_url}{movie_details_endpoint}', params=params)
    movie_details = movie_details_response.json()

    # Fetch keywords for each movie
    keywords_endpoint = f'/movie/{movie_id}/keywords'
    keywords_response = requests.get(f'{base_url}{keywords_endpoint}', params=params)
    keywords_data = keywords_response.json()

    # Extract required fields
    movie_data = {
        'id': movie_id,
        'genres': [genre['name'] for genre in movie_details['genres']],
        'keywords': [keyword['name'] for keyword in keywords_data['keywords']],
        'original_title': movie_details['original_title'],
        'overview': movie_details['overview'],
        'popularity': movie_details['popularity'],
        'vote_average': movie_details['vote_average'],
        'vote_count': movie_details['vote_count'],
        'title': movie_details['title']
        
    }

    movies_data.append(movie_data)

# Create a DataFrame from the collected movie data
movies_df = pd.DataFrame(movies_data)

# Save the DataFrame to a CSV file
movies_df.to_csv('movies_info.csv', index=False)

print("Movie data has been successfully stored in 'movies_info.csv'.")
```
